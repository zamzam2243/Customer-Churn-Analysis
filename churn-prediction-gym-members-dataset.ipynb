```python
col_f=["Age","Avg_Calories_Burned","Total_Weight_Lifted_kg","Visits_Per_Month"]
for col in col_f:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum()
```

```python
col_s=["Name"]
for col in col_s:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"\D", "", regex=True)
df["Phone_Number"] = df["Phone_Number"].fillna("0")
df["Phone_Number"] = df["Phone_Number"].astype("Int64")
```

```python
df["Last_Visit_Date"]=pd.to_datetime(df["Last_Visit_Date"],dayfirst=True,errors="coerce")
df["Join_Date"]=pd.to_datetime(df["Join_Date"],dayfirst=True,errors="coerce")
```

```python
da_columns=["Last_Visit_Date","Join_Date"]
for col in da_columns:
    df[col].fillna(pd.Timestamp('2000-01-01'), inplace=True)
```
